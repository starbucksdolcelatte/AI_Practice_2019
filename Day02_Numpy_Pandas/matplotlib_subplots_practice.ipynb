```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
```

```python
x = np.arange(20)
fig, ax = plt.subplots(1,2, figsize = (8,4))
ax[0].plot(x, x**2, label = 'y = x^2', 
           marker = 'o',
          color = 'green',
          linestyle = ':')
ax[0].plot(x, x, label = 'y = x',
          marker = '^',
          color = 'yellow')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[0].set_xlim(0,21)
ax[0].set_ylim(-100,500)

ax[0].legend(
    loc = 'upper left',
    shadow = True,
    fancybox = True,
    borderpad = 1
)
```

```python
y = np.random.randint(0,50,20)
ax[1].scatter(x, y, label = 'true',
             color = 'blue')
ax[1].plot(x, 3*x, label = 'pred',
          color = 'red')
ax[1].set_xlim(-5,35)
ax[1].set_ylim(-5,70)

ax[1].set_xlabel('rain')
ax[1].set_ylabel('dog_barks')

ax[1].legend(
    loc = 'upper right',
    shadow = True,
    fancybox = True,
    borderpad = 1
)
fig
```
